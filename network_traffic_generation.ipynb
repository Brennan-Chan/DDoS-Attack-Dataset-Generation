```python
# Re-run after kernel reset to regenerate the script output
import random
import time
import socket
import os
import requests
import pandas as pd

# Configuration
HOST = "http://fakenews.local:5000"

# we are going to pull from a csv for the following 
FAKE_USERNAMES = ["alice", "bob", "charlie", "daisy", "eva"] 
COMMENTS = ["great article", "fake news", "interesting!", "needs sources", "lol"]

# only items on the website
ITEMS = ["Mug", "Hat", "Sticker"]
IPS = [f"10.100.115.{i}" for i in range(10, 40)]  # Simulated IPs (avoiding VPN-style ranges)

# Create a fake session object
session = requests.Session()
session.verify = False

def simulate_http_traffic():
    username = random.choice(FAKE_USERNAMES)
    story_id = str(random.randint(1, 3))

    session.get(HOST + "/")
    session.get(HOST + "/stories")
    session.get(HOST + f"/story/{story_id}")
    session.post(HOST + "/login", data={"username": username, "password": "test"})
    session.post(HOST + f"/story/{story_id}", data={"comment": random.choice(COMMENTS)})
    session.get(HOST + f"/like/{story_id}")
    session.get(HOST + "/shop")
    session.post(HOST + "/shop/checkout", data={"item": random.choice(ITEMS), "qty": random.randint(1, 5)})
    session.get(HOST + "/api/data")
    session.post(HOST + "/api/data", json={"user": username, "action": "refresh"})

def simulate_tcp_raw():
    try:
        s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
        s.settimeout(2)
        s.connect(("fakenews.local", 5000))
        s.send(b"RAW TCP PROBE\r\n")
        s.close()
    except Exception:
        pass

def simulate_icmp():
    os.system("ping -c 2 fakenews.local > /dev/null")

def simulate_user_session():
    fake_ip = random.choice(IPS)
    print(f"\n[+] Simulating user from IP: {fake_ip}")

    simulate_icmp()
    simulate_http_traffic()
    simulate_tcp_raw()

# Run multiple sessions
for _ in range(10):  # Number of fake users
    simulate_user_session()
    time.sleep(random.uniform(1, 3))

```
