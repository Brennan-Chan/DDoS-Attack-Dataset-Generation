```python
import random
import pandas as pd

# Define percentages for each region (must sum to 1.0)
region_distribution = {
    "US (Comcast)": {"base": "73.162.144", "percentage": 0.35},
    "US (AT&T)": {"base": "99.112.8", "percentage": 0.25},
    "UK (BT)": {"base": "86.128.0", "percentage": 0.10},
    "Canada (Bell)": {"base": "142.114.0", "percentage": 0.10},
    "Australia (Optus)": {"base": "118.208.0", "percentage": 0.08},
    "Germany (Telekom)": {"base": "87.123.0", "percentage": 0.07},
    "Global Misc": {"base": "45.66.77", "percentage": 0.05}
}

TOTAL_IPS = 100  # Total number of IPs to generate

# Function to generate IPs in a subnet
def generate_cluster(base, count):
    return [f"{base}.{i}" for i in random.sample(range(1, 255), count)]

# Generate IPs based on percentages
ip_data = []
for region, config in region_distribution.items():
    count = int(config["percentage"] * TOTAL_IPS)
    cluster_ips = generate_cluster(config["base"], count)
    for ip in cluster_ips:
        ip_data.append({"Region": region, "IP": ip})

df_percent_ips = pd.DataFrame(ip_data)
df_percent_ips.head(20)
```

```python
# once we have some IPS we like save them to a CSV file
df_percent_ips.to_csv("ip_distribution_V1.csv", index=False)
```
